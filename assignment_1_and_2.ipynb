# **Assignment 2**

```python
def counter(x):
  N=sorted(set(x))
  D={}
  for i in N:
    D.update({i:x.count(i)})
  return D
```

```python
counter([3, 4, 2, 1, 5, 1, 2, 4, 3, 4])
```
